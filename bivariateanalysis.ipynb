```python
import pandas as pd
import numpy as np
import seaborn as sns
# https://www.kaggle.com/datasets/whenamancodes/students-performance-in-exams
df = pd.read_csv("exams.csv")
df.head()
```

<b> Categorical vs. Categorical Variables

```python
pd.crosstab(   df["race/ethnicity"], df["parental level of education"], margins=True,     values=df["math score"], aggfunc=pd.Series.count)
```

```python
sns.heatmap(pd.crosstab(   df["test preparation course"], df["parental level of education"], margins=False,     values=df["math score"], aggfunc=pd.Series.mean),
            cmap="YlGnBu", annot=True, cbar=False)
```

```python
pd.pivot_table(df,index =["gender","test preparation course"],aggfunc=[np.mean])
```

```python
pd.pivot_table(df,index =["parental level of education","gender"],aggfunc=[np.std])
```

<b> Categorical vs. Continuous Variables

```python
sns.barplot(x='test preparation course',y='math score',data=df)
```

```python
sns.countplot(x='gender',data=df)
```

```python
sns.boxplot(x='test preparation course',y='writing score',data=df,palette='rainbow')
# sns.swarmplot(x="day", y="total_bill", data=t)
```

```python
sns.violinplot(x='test preparation course',y='reading score',data=df,palette='rainbow')
```

```python
sns.stripplot(x='test preparation course',y='math score',data=df)
```

```python
sns.swarmplot(x='test preparation course',y='math score',data=df)
```
